```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import mean
from numpy import std
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_validate
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import SelectKBest
from sklearn.neighbors import KNeighborsClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
import sys
sys.path.append('./')

from helpers import functions as fn
```

```python
dataset = pd.read_csv('dataset_full.csv')
Y = pd.read_csv('emtab_tags.csv')
```

```python
dataset
```

```python
Y
```

```python
rfc = RandomForestClassifier(n_estimators=100)
svmc = svm.SVC(kernel='linear', C=1)
lrc = LogisticRegression(random_state=0)
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
fs = SelectKBest(score_func=f_classif)

X_selected = fs.fit_transform(dataset, Y)
```

```python
cols = fs.get_support(indices=True)
cols
```

```python
dataset_reduced = dataset.iloc[:,cols]
dataset_reduced
```

```python
rfc_scores_filter = cross_validate(rfc, dataset_reduced, Y['result'], scoring=('accuracy', 'f1', 'roc_auc'), cv=10, return_estimator =True)
```

```python
print("%0.3f accuracy with a standard deviation of %0.3f" % (rfc_scores_filter['test_accuracy'].mean(), rfc_scores_filter['test_accuracy'].std()))
print("%0.3f f1 score with a standard deviation of %0.3f" % (rfc_scores_filter['test_f1'].mean(), rfc_scores_filter['test_f1'].std()))
print("%0.3f roc_auc with a standard deviation of %0.3f" % (rfc_scores_filter['test_roc_auc'].mean(), rfc_scores_filter['test_roc_auc'].std()))
```

```python
svm_scores_filtered = cross_validate(svmc, dataset_reduced, Y['result'], scoring=('accuracy', 'f1', 'roc_auc'), cv=10, return_estimator =True)
```

```python
print("%0.3f accuracy with a standard deviation of %0.3f" % (svm_scores_filtered['test_accuracy'].mean(), svm_scores_filtered['test_accuracy'].std()))
print("%0.3f f1 score with a standard deviation of %0.3f" % (svm_scores_filtered['test_f1'].mean(), svm_scores_filtered['test_f1'].std()))
print("%0.3f roc_auc with a standard deviation of %0.3f" % (svm_scores_filtered['test_roc_auc'].mean(), svm_scores_filtered['test_roc_auc'].std()))
```

```python
lrc_scores_reduced = cross_validate(lrc, dataset_reduced, Y['result'], scoring=('accuracy', 'f1', 'roc_auc'), cv=10, return_estimator =True)
```

```python
print("%0.3f accuracy with a standard deviation of %0.3f" % (lrc_scores_reduced['test_accuracy'].mean(), lrc_scores_reduced['test_accuracy'].std()))
print("%0.3f f1 score with a standard deviation of %0.3f" % (lrc_scores_reduced['test_f1'].mean(), lrc_scores_reduced['test_f1'].std()))
print("%0.3f roc_auc with a standard deviation of %0.3f" % (lrc_scores_reduced['test_roc_auc'].mean(), lrc_scores_reduced['test_roc_auc'].std()))
```

```python
knn_scores_reduced = cross_validate(knn, dataset_reduced, Y['result'], scoring=('accuracy', 'f1', 'roc_auc'), cv=10, return_estimator =True)
```

```python
print("%0.3f accuracy with a standard deviation of %0.3f" % (knn_scores_reduced['test_accuracy'].mean(), knn_scores_reduced['test_accuracy'].std()))
print("%0.3f f1 score with a standard deviation of %0.3f" % (knn_scores_reduced['test_f1'].mean(), knn_scores_reduced['test_f1'].std()))
print("%0.3f roc_auc with a standard deviation of %0.3f" % (knn_scores_reduced['test_roc_auc'].mean(), knn_scores_reduced['test_roc_auc'].std()))
```
