```python
from scipy.stats import mannwhitneyu
from statannotations.Annotator import Annotator
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
def create_boxplot(data, x="Data", y="Accuracy", pairs=[("Original", "FSC")], alg_name="K Nearest Neighbors", pdf_name="knn_acc.pdf"):
    x = x
    y = y
    pairs=pairs
    a4_dims = (5, 7)
    fig, ax = plt.subplots(figsize=a4_dims)
    ax.set_title(alg_name, fontsize=19)
    sns.boxplot(data=data, x=x, y=y , width=0.25)
    annotator = Annotator(ax, pairs, data=data, x=x, y=y)
    annotator.configure(test='Mann-Whitney', text_format='star', loc='inside')
    annotator.apply_and_annotate()
    plt.savefig(pdf_name, dpi=150, bbox_inches='tight')
```
